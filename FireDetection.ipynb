```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model

import numpy as np
import scipy as sp
import cv2
from keras.preprocessing import image

import matplotlib.pyplot as plt
%matplotlib inline


classifier = Sequential()
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Flatten())
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)
valid_datagen = ImageDataGenerator(rescale = 1./255)
training_set = train_datagen.flow_from_directory('FireData/train',
                                                 target_size = (64, 64),
                                                 batch_size = 32, 
                                                 color_mode="rgb",
                                                 class_mode = 'binary', 
                                                 shuffle=True,
                                                 seed=42)
valid_set = valid_datagen.flow_from_directory(
    'FireData/test',
    target_size=(64, 64),
    color_mode="rgb",
    batch_size=32,
    class_mode="binary",
    shuffle=True,
    seed=42
)
```

```python
classifier.fit_generator(training_set,
                         steps_per_epoch = 500,
                         epochs = 1000,
                         validation_data = valid_set,
                         validation_steps = 200)

classifier.save('fireclassifier.h5')
```

```python

def predict(imgage_path):
    test_image = image.load_img(img_path, target_size = (64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = classifier.predict(test_image)
    print(training_set.class_indices)
    if result[0][0] == 1:
        prediction = 'notfire'
        return False
    else:
        prediction = 'fire'
        return True

def predict_part(inp_arr_image):
    test_image = cv2.resize(inp_arr_image, (64,64))
    test_image = np.expand_dims(test_image, axis = 0)
    result = classifier.predict(test_image)
    training_set.class_indices
    if result[0][0] == 1:
        prediction = 'notfire'
        return False
    else:
        prediction = 'fire'
        return True

def plot_image(img_array):
    img = np.array(img_array, np.int32)
    plt.imshow(img)
    

def get_cells_img(np_arr_img, n=64): # considers all n X n grids
    sub_imgs = []
    for row in range((np_arr_img.shape[0]//n)):
        for col in range((np_arr_img.shape[1]//n)):
            c_0 = col*n
            c_1 = min((c_0+n), np_arr_img.shape[1])
            r_0 = row*n
            r_1 = min((r_0+n), np_arr_img.shape[0])
    #         print(c_0, c_1, " | ", r_0, r_1)
            sub_imgs.append(np_arr_img[r_0:r_1, c_0:c_1,: ])
    return sub_imgs

def get_img_array(path):
    img = image.load_img(path)
    return image.img_to_array(img)

def plot_images(img, labels, nrows, ncols):
    fig, axes = plt.subplots(nrows, ncols)
    print(len(axes.flat))
    for i, ax in enumerate(axes.flat): 
        ax.imshow(img[i])
        ax.set_xticks([]); ax.set_yticks([])
        ax.set_title(labels[i])

def count_subimgs(img_arr_pred):
    fire_cnt = 0
    for s in img_arr_pred:
            if s:
                fire_cnt += 1

    no_fire_cnt = len(img_arr_pred) - fire_cnt

    print("fire: ", fire_cnt, "non-fire:", no_fire_cnt)
```

```python
test_img_fire = image.load_img('FireData/vald/fire1.jpg')
test_img_fire = image.img_to_array(test_img_fire)
print(test_img_fire.shape)
plot_image(test_img_fire)
```

```python
test_img_no_fire = image.load_img('FireData/vald/no_fire.jpeg')
test_img_no_fire = image.img_to_array(test_img_no_fire)
print(test_img_no_fire.shape)
plot_image(test_img_no_fire)
```

```python
test_img_fire_arr = get_img_array('FireData/vald/fire1.jpg')
test_img_fire_arr = image.img_to_array(test_img_fire_arr)
test_img_fire_arr = cv2.resize(test_img_fire_arr, (640, 640) )
print(test_img_fire_arr.shape)
plot_image(test_img_fire_arr)
test_img_fire_arr_cells = get_cells_img(test_img_fire_arr, n=128)
test_img_fire_arr_pred = [predict_part(img) for img in test_img_fire_arr_cells]
```

```python
test_img_no_fire_arr = get_img_array('FireData/vald/no_fire.jpeg')
test_img_no_fire_arr = cv2.resize(test_img_no_fire_arr, (640, 640) )
print(test_img_no_fire_arr.shape)
plot_image(test_img_no_fire_arr)
test_img_no_fire_arr_cells = get_cells_img(test_img_no_fire_arr, n=128)
test_img_no_fire_arr_pred = [predict_part(img) for img in test_img_no_fire_arr_cells]
```

```python
count_subimgs(test_img_fire_arr_pred)
fire_sub_imgs = [np.array(img, np.int32) for img in test_img_fire_arr_cells]
plot_images(fire_sub_imgs, test_img_fire_arr_pred, 5, 5)
```

```python

count_subimgs(test_img_no_fire_arr_pred)
no_fire_sub_imgs = [np.array(img, np.int32) for img in test_img_no_fire_arr_cells]
plot_images(no_fire_sub_imgs, test_img_no_fire_arr_pred, 5, 5)
```
